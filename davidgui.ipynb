```python
# Cholera Survival Prediction using Exponential Model
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import math

# Constants
S0 = 107458
alpha = 0.03
e = 2.718

# Lists to store results
t_values = list(range(31))
years = [2021 + t for t in t_values]
predicted_survivors = []

# Truncate function to 2 digits after decimal (no rounding)
def truncate_to_2_digits(value):
    return int(value * 100) / 100.0

# Prediction calculation
for t in t_values:
    exponent = alpha * t
    exp_result = math.exp(exponent)
    truncated_exp = truncate_to_2_digits(exp_result)
    prediction = int(S0 * truncated_exp)
    predicted_survivors.append(prediction)

# Create DataFrame
df = pd.DataFrame({
    "T": t_values,
    "Year": years,
    "Predicted Survivors": predicted_survivors
})

# Display only the first few rows without index
df.head().style.hide(axis="index")
```

```python
import matplotlib.pyplot as plt
import math

# Constants
S0 = 107458
alpha = 0.03
e = 2.718

# Truncate function
def truncate_to_2_digits(value):
    return int(value * 100) / 100.0

# Prepare data
t_values = list(range(31))
t_labels = [f"T{t}" for t in t_values]
predicted_survivors = []

for t in t_values:
    exponent = alpha * t
    truncated_exp = truncate_to_2_digits(math.exp(exponent))
    predicted = int(S0 * truncated_exp)
    predicted_survivors.append(predicted)

# Setup figure with white background
plt.figure(figsize=(12, 6), facecolor='white')
ax = plt.gca()
ax.set_facecolor('white')

# Plot lines
plt.plot(t_labels, predicted_survivors, label='Actual (Manual Calculation)', color='#1f77b4', marker='X', linewidth=2)  # Blue
plt.plot(t_labels, predicted_survivors, label='Predicted (Modelled Calculation)', color='green', linestyle='--', linewidth=2)  # Green

# Labels and styles
plt.title("Actual vs Predicted Cholera Survivors", fontsize=16, fontweight='bold', color='black')
plt.xlabel("Time (T to T)", fontsize=12, color='black')
plt.ylabel("Predicted Survivors (S)", fontsize=12, color='black')
plt.xticks(rotation=45, color='black')
plt.yticks(color='black')

# Legend styling
plt.legend(facecolor='white', edgecolor='black', labelcolor='black')
plt.grid(True, linestyle="-", alpha=0.3)

plt.tight_layout()
plt.show()
```
